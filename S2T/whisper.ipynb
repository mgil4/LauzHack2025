```python
model = whisper.load_model("small")

# Carpeta de videos
UPLOADS_DIR = r".\server\uploads"

# Obtener todos los .webm
video_paths = [
    os.path.join(UPLOADS_DIR, f)
    for f in os.listdir(UPLOADS_DIR)
    if f.lower().endswith((".webm", ".mp4", ".mov", ".avi"))
]

# Carpeta donde guardar textos
OUTPUT_DIR = r".\transcripts"
os.makedirs(OUTPUT_DIR, exist_ok=True)

for video_path in video_paths:
    print(f"\nðŸŽ¥ Transcribiendo: {video_path}")

    result = model.transcribe(video_path, language="en")
    transcript = result["text"]

    base_name = os.path.basename(video_path)
    name_no_ext = os.path.splitext(base_name)[0]
    transcript_file = os.path.join(OUTPUT_DIR, f"{name_no_ext}.txt")

    print("TranscripciÃ³n:\n", result["text"])
    
    # Guardar transcripciÃ³n
    with open(transcript_file, "w", encoding="utf-8") as f:
        f.write(transcript)

    print(f"ðŸ“„ Guardado en: {transcript_file}")

print("\nâœ” Todos los videos han sido transcritos.")
```
